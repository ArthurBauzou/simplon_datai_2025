```python
#Tracer un histogramme à partir de données aléatoires générées par NumPy.

data = np.random.normal(0,40,1000)
plt.hist(data, bins=12, linewidth=0.8, edgecolor="white");
```

```python
# Tracer un diagramme à barres à partir des données suivantes : 
# categories = ['A', 'B', 'C', 'D', 'E'], valeurs = [25, 30, 15, 20, 10]

categories = np.array(['A', 'B', 'C', 'D', 'E'])
valeurs = [25, 30, 15, 20, 10]

plt.bar(categories,valeurs, color='purple')
```

```python
# Tracer un nuage de points à partir de données aléatoires générées par NumPy.

x = np.random.randn(1000)
y = np.random.randn(1000)


plt.plot(x,y,'o')
plt.show()
plt.scatter(x,y)
```

```python
# Tracer une courbe sinusoïdale en utilisant NumPy.

x = np.arange(-20,21,0.2)

plt.plot(x,np.sin(x))
plt.ylim(-2.5,2.5)
```

```python
# Tracer deux courbes sur le même graphe à partir des données suivantes :
# x = np.linspace(0, 10, 100)
# y1 = np.sin(x)
# y2 = np.cos(x)

x = np.linspace(0, 10, 50)

plt.plot(x, np.sin(x), '--')
plt.plot(x, np.cos(x), color="orange")
plt.ylim(-2,2)
plt.grid()
plt.axis('scaled')
plt.show()
```

```python
# Tracer un diagramme en secteurs (camembert) à partir des données suivantes :
# categories = ['A', 'B', 'C', 'D'], valeurs = [30, 25, 15, 20]

categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]

plt.pie(
    valeurs, 
    labels=categories, 
    startangle=90,
    colors=["gray", "pink", "beige", "peachpuff"],
    explode=[0,0,0.1,0]
    )

plt.show()
```

```python
# Tracer une courbe polynomiale à partir des données suivantes :
# x = np.linspace(-5, 5, 100)
# y = 3 * x**3 + 2 * x**2 - 8 * x + 5

x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

plt.plot(x,y)
```

```python
# Tracer un diagramme en barres empilées à partir des données suivantes :
# categories = ['A', 'B', 'C', 'D']
# valeurs1 = [10, 15, 20, 25], valeurs2 = [5, 10, 15, 20]

categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25]
valeurs2 = [5, 10, 15, 20]

plt.barh(categories, valeurs1)
plt.barh(categories, valeurs2);
```

```python
# Tracer un graphique en boîte (boxplot) à partir des données suivantes :
# data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]

data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]

plt.boxplot(data, notch=True);
        
```

```python
# dessiner des cercles concentriques avec une fonction mathématique

def concentriques(number, precision=40):
    if number <= 1 or number > 9:
        print('nope')
        return 0
    
    x = np.arange(0,2*np.pi,1/precision)
    colors = ['#33114a', '#6d143d', '#9c1733', '#cb1a28', '#fd1d1d', '#fe4c31', '#fe7743', '#ff9c53', '#ffbd61']
    for i in range(1,number+1):
        plt.plot(
            np.cos(x)/(10+i),
            np.sin(x)/(10+i),
            '-', 
            color=colors[i-1])
    plt.axis('scaled')
    plt.show()

concentriques(8, 100)
```
