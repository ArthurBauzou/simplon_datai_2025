```python
import random as rd

items = [
    'un tube de super glue',
    'un gros jambon espagnol',
    'un couteau japonais',
    'une pagaie fluo de chez decathlon',
    'un marteau de tape-taupe',
    'une veste moche portée au met gala',
    'du thon en boite bio, ethique, équitable et même carrement vegan',
    'deux bichons frisés',
    'une chevalière de biker',
    'un cochon truffier'
    ]

superitems = [
    'un tigre vert au pelage soyeux',
    'une épée maudite mais très sympa',
    'des grenades en caoutchouc qui rebondissent super bien',
    'une bouteille contenant tout le coca-cola',
    'des collants mega resistants qui ne filent jamais',
    'un majordome avec un accent britannique',
    'une voiture qui roule sur les murs',
    'le dernier tome de song of fire and ice en exclu'
]

superlatif = ['bien', 'particulièrement', 'très']

perks = [
    ('cupide', {'int': 1, 'cha': -1}),
    ('ambitieux', {'cha': 1}),
    ('bodybuildé', {'dex': -1, 'for': 2}),
    ('curieux', {'int': 1}),
    ('bavard', {'cha': 1, 'int': -1}),
    ('anxieux', {'dex': -1, 'int': 1}),
    ('habile', {'dex': 2, 'for': -1}),
    ('sale', {'cha': -2}),
    ('svelte', {'cha': 1, 'dex': 1})
]

stat_descr = {
    '1' : 'complètement nul',
    '2' : 'médiocre et décevant',
    '3' : 'sympa, mais pas ouf non plus',
    '4' : 'plutot correct',
    '5' : 'bien au dessus de la moyenne',
    '6' : 'très très fort',
    '7' : 'la classe mondiale, top of the pops',
    'cha' : 'charisme',
    'dex' : 'dextérité',
    'for' : 'force',
    'int' : 'intelligence',
    'cha_t' : ['son bagout et sa belle gueule','son évidente aura de leader'],
    'dex_t' : ['sa souplesse et sa grace','sa foudroyante rapidité'],
    'for_t' : ['son inébranlable volonté','son corps puissant et musculeux'],
    'int_t' : ['sa malice et son sens de l’observation','son implacable et rigoureux esprit logique'],
    'cha_d' : ['sa voix de goéland insupportable','sa vulgarité et ses manières de phacochère'],
    'dex_d' : ['sa démarche raide et incertaine', 'son insupportable lenteur'],
    'for_d' : ['son corps maigre et maladif', 'sa couardise et son manque de conviction'],
    'int_d' : ['son premier degré confinant à l’idiotie', 'son inculture et ses préjugés']
}

#––––––– DUDE –––––––#

class Dude:
    def __init__(self, name='Sans-Nom') -> None:
        self.name = name
        self.stats = {
            'for': rd.choice(range(5))+1,
            'dex': rd.choice(range(5))+1,
            'int': rd.choice(range(5))+1,
            'cha': rd.choice(range(5))+1
        }
        # perks
        self.perk = ''
        perk = rd.choice(perks)
        self.apply_perk(perk)
        self.stuff = []
        self.stuff.append(rd.choice(items))

    def apply_perk(self, perk):
        self.perk = perk[0]
        for k,v in perk[1].items():
            self.stats[k] += v
            if self.stats[k] < 1 : self.stats[k] = 1

    def rename(self, name):
        self.name = name

    def drop_stuff(self, item):
        if item not in self.stuff:
            return False
        else:
            self.stuff.remove(item)
            return True
        
    def add_stuff(self, item):
        if len(self.stuff) >= self.stats['for']:
            return False
        else:
            self.stuff.append(item)
            return True
        
    def get_talent(self):
        score = max(self.stats.values())
        stat = []
        for k,v in self.stats.items():
            if v == score : stat.append(k)

        stat = rd.choice(stat)
        talent = rd.choice(stat_descr[stat+'_t'])
        scoretext = stat_descr[str(score)]

        return {
            'stat': stat_descr[stat],
            'descr': talent,
            'score': score,
            'score_d': scoretext
        }
    
    def get_weakness(self):
        score = min(self.stats.values())
        stat = []
        for k,v in self.stats.items():
            if v == score : stat.append(k)

        stat = rd.choice(stat)
        defaut = rd.choice(stat_descr[stat+'_d'])
        scoretext = stat_descr[str(score)]

        return {
            'stat': stat_descr[stat],
            'descr': defaut,
            'score': score,
            'score_d': scoretext
        }
    
    def get_level(self):
        return f'DUDE LVL : {sum(self.stats.values())}'


#––––––– HERO –––––––#

class Hero(Dude):
    def __init__(self, name='Sans-Nom', buff=None) -> None:
        super().__init__(name)

        self.stuff.append(rd.choice(superitems))

        self.superstat = None
        if buff == None : buff = rd.choice(['cha', 'int', 'for', 'dex'])
        self.stat_buff(buff)


    def stat_buff(self, stat):
        self.stats[stat] += 3
        if self.stats[stat] > 7 :
            self.superstat = stat


    def get_talent(self):
        res = {}
        if self.superstat : 
            tempon = self.stats[self.superstat]
            self.stats[self.superstat] = 0
            res = super().get_talent()
            res['score'] = tempon
            res['score_d'] = 'terrifiant et surhumain'
            self.stats[self.superstat] = tempon
        else :
            res = super().get_talent()

        return res
    
    def get_level(self):
        return f'HERO LVL : {sum(self.stats.values())}'


#––––––– General ––––––––#

def print_special(talent, special):
    intro = 'Son talent réside dans' if talent else 'Son plus grand défaut est'
    a = f'  > {intro} {special["descr"]},'
    b = f'\n    avec un score de {special["score"]} en {special["stat"]},'
    c = f' ce qui est {special["score_d"]}.'
    print(a + b + c)

def print_stuff(name, stuff):
    rd.shuffle(stuff)
    msg = []
    msg.append(f'  > {name} possède {stuff[0]}')
    if len(stuff) > 1:
        msg.append(f' et aussi {",".join(stuff[1:])}')

    print(''.join(msg))

def print_perso(name):
    perso = rd.choice([Hero, Dude])(name)
    print(f'\n––––  {perso.name.upper()}  ––––')
    print(perso.get_level())
    print(f'  > {perso.name} est {rd.choice(superlatif)} {perso.perk}')
    print_stuff(perso.name, perso.stuff)
    print_special(True, perso.get_talent())
    print_special(False, perso.get_weakness())
    

for n in ['Alex', 'Stone', 'Stanley', 'Yu', 'Flubber']:
    print_perso(n)


```
