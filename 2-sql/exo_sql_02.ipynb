```python
# importer les données depuis le document excel «utilisateurs»

import numpy as np
import pandas as pd

util_df = pd.read_excel("./utilisateurs.xlsx")

util_df.drop(['TYPE', 'relecteur'], axis=1, inplace=True)
util_df['email2'] = 'undefined'

print(util_df.isna().apply(sum))
# > aucune valeur de manque

ids = util_df.groupby('id').count().sort_values('nom', ascending=False)
# print(ids)
# > Les identifiants sont bien tous uniques

nprof = util_df.groupby('profession').count().sort_values('nom', ascending=False)
# print(nprof)
# > créer une table des professions avec 5 entrée

nvil = util_df.loc[:,['ville', 'nom']].groupby('ville').count().sort_values('nom', ascending=False)
# print(nvil)
# > 505 villes

npays = util_df.loc[:,['pays', 'nom']].groupby('pays').count().sort_values('nom', ascending=False)
# print(npays)
# > 240 Pays différents

ngenr = util_df.loc[:,['genre', 'nom']].groupby('genre').count().sort_values('nom', ascending=False)
# print(ngenr)
# > 3 genres


```

```python
# Objectif : vérifier que les villes sont bien toutes associées avec le même pays

# générer un tableau d’occurences de chaque villes
nvil.reset_index(inplace=True)
nvil.columns = ['ville', 'n']
# on ne garde que celles qui apparaissent plus d’une fois
nvil = nvil[nvil["n"] > 1]

# affichage des pays correspondant en fonction des villes
for i in range(nvil.shape[0]) :
    payvil = util_df[util_df['ville'] == nvil['ville'][i]]
    print(payvil.loc[:,['ville', 'pays']])
# > On peut voir que les pays et les villes n’ont rien à voir dans nos données
```
